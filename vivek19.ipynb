# bar plot in seaborn liberary

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = sns.load_dataset("penguins")
```

```python
df1
```

```python
df1["island"]
```

```python
sns.barplot(x = df1.island, y = df1.bill_length_mm)
plt.show()
```

# ak tarika or hota hai isa draw karna ka

```python
sns.barplot(x = "island" , y = "bill_length_mm" ,data = df1)
plt.show()
```

# ORDER

```python
order1 = ["Dream" , "Torgersen" , "Biscoe"]
sns.barplot(x= "island" , y = "bill_length_mm", data =df1 , hue = "sex" , order = order1)
plt.show()
```

# agar huma male femail ka order bhi change karna ho tho

```python
order1 = ["Dream" , "Torgersen" , "Biscoe"]
sns.barplot(x= "island" , y = "bill_length_mm", data =df1 , hue = "sex" , 
            order = order1, hue_order = ["Female" , "Male"])
plt.show()
```

```python
order1 = ["Dream" , "Torgersen" , "Biscoe"]
sns.barplot(x= "island" , y = "bill_length_mm", data =df1 , color = "green")
plt.show()
```

# saturation parameter

# colour dark ya light karna ka liya

```python
sns.barplot(x = "island" , y = "bill_length_mm" , data = df1 , hue = "sex" , 
            palette = "prism",saturation = 0.3)
plt.show()
```

```python
sns.barplot(x = "island" , y = "bill_length_mm" , data = df1 , hue = "sex" , 
            palette = "prism",saturation = 0.50)
plt.show()
```

# Histogram in seaborn liberary

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

df1

```python
sns.displot(df1["flipper_length_mm"])
plt.show()
```

# ab yha dhak raha hai yha proper gab nahi hai 170 _ 180 ....... ya karna ka liya asa karta hai

# bins parameter use kara ga

```python
sns.displot(df1['flipper_length_mm'], bins = [170,180,190,200,210,220,230,240])
plt.show()
```

# color

```python
sns.displot(df1['flipper_length_mm'], bins = [170,180,190,200,210,220,230,240],
           color = "g")
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'], bins = [170,180,190,200,210,220,230,240],
           color = "r")
plt.show()
```

# kernal density = kde

```python
sns.displot(df1['flipper_length_mm'], bins = [170,180,190,200,210,220,230,240,250,260] , kde = True)
plt.show()
```

# agar kde nahi chiya tho ------ kde = False ----- kar data hai

```python
sns.displot(df1['flipper_length_mm'], bins = [170,180,190,200,210,220,230,240,250,260] , kde = False)
plt.show()
```
