# width ko apna according set karna

```python
# color ka code bhi dal sakta hai
x = np.array(['a' ,'b' ,'c' ,'d'])
y = np.array([30,18,11,19])
             
plt.bar(x,y, color = "#4CAF50", width = 0.6 )
plt.show()
```

```python
# color ka code bhi dal sakta hai
x = np.array(['a' ,'b' ,'c' ,'d'])
y = np.array([30,18,11,19])
             
plt.bar(x,y, color = "#4CAF50" , width = 0.1 )
plt.show()
```

# agar hum graoh ko horizontlly represent kar raha hai tho uski hight kasa kam karta hai

# scatter plt

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]

plt.scatter(x,y)
plt.show()
```

# tital dana ka liya

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]

plt.title("scatter plot")
plt.scatter(x,y)
plt.show()
```

# Agar x or y ka label dana hai tho ya karta hai

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]

plt.title("scatter plot")
plt.xlabel("month")
plt.ylabel("number")
plt.scatter(x,y)
plt.show()
```

# colour change karna ka liya

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]

plt.title("scatter plot")
plt.xlabel("month")
plt.ylabel("number")
plt.scatter(x,y)
plt.scatter(x,y, color = "red")
plt.show()
```

# Agar sab ka olour alag alag karna chata hai tho ya karta hai

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]

plt.title("scatter plot")
plt.xlabel("month")
plt.ylabel("number")

c = ['red' , 'blue' , 'green' , 'orange' , 'magenta' , 'black']
plt.scatter(x,y, color = c)
plt.show()
```

# size change karna ka liya

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]

plt.title("scatter plot")
plt.xlabel("month")
plt.ylabel("number")

c = ['red' , 'blue' , 'green' , 'orange' , 'magenta' , 'black']
plt.scatter(x,y, color = c , s = 150)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]

plt.title("scatter plot")
plt.xlabel("month")
plt.ylabel("number")

c = ['red' , 'blue' , 'green' , 'orange' , 'magenta' , 'black']
plt.scatter(x,y, color = c, s = 1000)
plt.show()
```
