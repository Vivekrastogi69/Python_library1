```python
var1 = [1,2,3,4,5,6,7,8]
var2 = [8,7,6,5,4,3,2,1]

import pandas as pd
df = pd.DataFrame({"var1":var1,"var2":var2})

sns.lineplot(x = "var1" , y = "var2" , data = df)
plt.show()
```

```python
df1 = sns.load_dataset("penguins")
```

```python
df1
```

```python
sns.lineplot(x = "bill_length_mm" , y = "bill_depth_mm" , data =df1)
plt.show()
```

# ab agar upar wala data ko mai gender wise devide karu tho

```python
sns.lineplot(x = "bill_length_mm" , y = "bill_depth_mm" , data =df1, hue = "sex")
plt.show()
```

# agar colour changekarna chat hia tho

```python
sns.lineplot(x = "bill_length_mm" , y = "bill_depth_mm" , data =df1, hue = "sex" , palette = "rocket_r")
plt.show()
```

# Agar huma line ko mark karna ho thpo asa karta hai

```python
sns.lineplot(x = "bill_length_mm" , y = "bill_depth_mm" , data =df1, hue = "sex" ,style = "sex" ,
             palette = "rocket_r", markers = ['o' , '>'])
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm" , y = "bill_depth_mm" , data =df1, hue = "sex" , style = "sex" , markers = ["o", ">"])
plt.show()
```
