# matplotlib - piechart

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,50]
y = ['english', 'hindi' , 'maths',
     'scientce']
plt.pie(x)
plt.title("growth in subjects")
plt.show()
```

# agar huma show karna hai kon kiska hai tho label data hai jasa nicha diya hai

```python
x = [10,20,30,50]
y = ['english', 'hindi' , 'maths',
     'scientce']
plt.pie(x, labels = y)
plt.title("growth in subjects")
plt.show()
```

# colour change karna ka liya

```python
x = [10,20,30,50,32]
y = ['english', 'hindi' , 'maths',
     'scientce','social']
c = ['yellow' , 'black' , 'magenta' , 'brown','orange']
plt.pie(x, labels = y , colors = c)
plt.title("growth in subjects")
plt.show()
```

# agar hum side mai ak label bana chata hai tho asa karta hai   plt.legend()   use karta hai

```python
x = [10,20,30,50,32]
y = ['english', 'hindi' , 'maths',
     'scientce','social']
c = ['yellow' , 'black' , 'magenta' , 'brown','orange']
plt.pie(x, labels = y , colors = c)
plt.title("growth in subjects")
plt.legend()
plt.show()
```
